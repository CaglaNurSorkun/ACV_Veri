```python
# Her ilçe için benzersiz müşteri sayılarını hesapla
hamamozu_musteri_sayisi = df_tahakkuk['sozlesme_hesap_no'].nunique()
gumushacikoy_musteri_sayisi = df_tahakkuk_1['sozlesme_hesap_no'].nunique()
goynucek_musteri_sayisi = df_tahakkuk_2['sozlesme_hesap_no'].nunique()

print(f"Hamamözü Müşteri Sayısı: {hamamozu_musteri_sayisi}")
print(f"Gümüşhacıköy Müşteri Sayısı: {gumushacikoy_musteri_sayisi}")
print(f"Göynücek Müşteri Sayısı: {goynucek_musteri_sayisi}")
```

```python
karsilastirma_verisi = {
    'İlçe': ['Hamamözü', 'Gümüşhacıköy', 'Göynücek'],
    'Müşteri Sayısı': [hamamozu_musteri_sayisi, gumushacikoy_musteri_sayisi, goynucek_musteri_sayisi]
}

df_karsilastirma = pd.DataFrame(karsilastirma_verisi)
# Yüzdesel payı da ekleyelim
total = df_karsilastirma['Müşteri Sayısı'].sum()
df_karsilastirma['Pay (%)'] = (df_karsilastirma['Müşteri Sayısı'] / total * 100).round(2)

print(df_karsilastirma)
```

```python
# Üç ilçenin tahakkuk verilerini alt alta birleştir
df_tum_tahakkuk = pd.concat(
    [df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], 
    axis=0, # Satır bazlı (alt alta) birleştirme
    ignore_index=True # İndeksleri 0'dan başlayarak yeniden düzenle
)
```

```python
df_tum_tahakkuk.head()
```

```python
# Ayrı ayrı satır sayıları
satir_hamamozu = len(df_tahakkuk)      # 124,818
satir_gumushacikoy = len(df_tahakkuk_1) # 765,657
satir_goynucek = len(df_tahakkuk_2)    # 295,223

toplam_beklenen = satir_hamamozu + satir_gumushacikoy + satir_goynucek
gerceklesen_toplam = len(df_tum_tahakkuk)

print(f"Beklenen Toplam Kayıt: {toplam_beklenen}")
print(f"Gerçekleşen Toplam Kayıt: {gerceklesen_toplam}")
```

```python
eksik_veriler = df_tum_tahakkuk['kwh'].isna().sum()
print(f"kwh sütunundaki eksik değer sayısı: {eksik_veriler}")
```

```python
negatif_veriler = df_tum_tahakkuk[df_tum_tahakkuk['kwh'] < 0]
print(f"Negatif tüketim kayıt sayısı: {len(negatif_veriler)}")
print(negatif_veriler['kwh'].describe()) # Negatiflerin istatistiksel özeti
```

```python

print("mean: ",df_tum_tahakkuk['kwh'].mean())
print("median: ",df_tum_tahakkuk['kwh'].median())
print("std: ",df_tum_tahakkuk['kwh'].std())
print("min: ",df_tum_tahakkuk['kwh'].min())
print("max: ",df_tum_tahakkuk['kwh'].max())
Q1 = df_tum_tahakkuk['kwh'].quantile(0.25)
Q3 = df_tum_tahakkuk['kwh'].quantile(0.75)
IQR = Q3 - Q1
print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
ust_sinir = Q3 + 1.5 * IQR
alt_sinir = Q1 - 1.5 * IQR

uc_degerler = df_tum_tahakkuk[(df_tum_tahakkuk['kwh'] > ust_sinir) | (df_tum_tahakkuk['kwh'] < alt_sinir)]
print(f"Tespit edilen aşırı uç değer sayısı: {len(uc_degerler)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
sns.boxplot(y=df_tum_tahakkuk['kwh'])
plt.title('kWh Tüketim Dağılımı (Odaklanmış Görünüm)')
plt.grid(True)
plt.show()
```

```python
# 'Hesap Sınıfı'na göre grupla ve kwh sütunu için istatistikleri hesapla
ozet_tablo = df_tum_tahakkuk.groupby('Hesap Sınıfı')['kwh'].agg(['mean', 'median', 'std']).reset_index()

# Sütun isimlerini daha anlaşılır hale getir
ozet_tablo.columns = ['Hesap Sınıfı', 'Ortalama (kWh)', 'Medyan (kWh)', 'Standart Sapma']

# Tabloyu ortalama tüketime göre büyükten küçüğe sırala 
ozet_tablo = ozet_tablo.sort_values(by='Ortalama (kWh)', ascending=False)

# Sonucu göster
print(ozet_tablo)
```
